```
#!/usr/bin/env python

import scipy
import numpy
import netCDF4
import csv

from numpy import arange, dtype
```

```
#Declare empty arrays

v1 = []
v2 = []
v3 = []
v4 = []
v5 = []
v6 = []
v7 = []
```

```
# Open csv file and declare variable for arrays for each heading

df = open('JUB_CURRENT.csv', 'r').readlines()
def str_to_float(str):
    try:
        number = float(str)
    except ValueError:
        number = 0.0
# you can assign an appropriate value instead of 0.0 which suits your requirement
    return number
for line in df[1:]:
    fields = line.split(',')
    v1.append(fields[0]) #time_utc
    v2.append(fields[1])#StationName
    v3.append(str_to_float(fields[2]))#Lat
    v4.append(str_to_float(fields[2]))#Lon
    v5.append(str_to_float(fields[3])) #current_speed
    v6.append(str_to_float(fields[3]))#current_direction
    v7.append(fields[4])#visibility_nm
#more variables included but this is just an abridged list
print (v1)
print (v2)
print (v3)
print (v4)
print (v5)
print (v6)
print (v7)
```

```
#convert to netcdf4 framework that works as a netcdf

ncout = netCDF4.Dataset('JUB_CURRENT.nc','w') 
```

```
# latitudes and longitudes. Include NaN for missing numbers

lats_out = 0 + 5.0*arange(v3,dtype='float32')
lons_out = 0 + 5.0*arange(v4,dtype='float32')
```

```
# create the lat and lon dimensions.

ncout.createDimension('latitude',v3)
ncout.createDimension('longitude',v4)
```
